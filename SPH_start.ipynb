```python
from SPHpy import *
```

### initialize stellar cloud with random positions and stationary velocities

```python
pos, vel, N = positions(N = 1000)
```

### run simulation for 400 timesteps

```python
position, densities = main_SPH(pos, vel, N, 2, .75, .1, 1, 1, 400, .04)
```

### plot several timesteps to see if stellar cloud collapsed into equilibrium state

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# default fig and camera viewing angles
cmap = 'plasma'

index1 = N*0 + N
index2 = N*100 + N
index3 = N*200 + N
index4 = N*350 + N
cval = np.minimum((densities-3)/3,1).flatten()
cval1 = np.minimum((densities[:index1]-3)/3,1).flatten()
cval2 = np.minimum((densities[N*100:index2]-3)/3,1).flatten()
cval3 = np.minimum((densities[N*200:index3]-3)/3,1).flatten()
cval4 = np.minimum((densities[N*350:index4]-3)/3,1).flatten()

# define figure
fig, axs = plt.subplots(2, 2, figsize = (30,30))

# formatting
plt.rcParams['font.family'] = 'sans-serif' # set font

a = axs[0,0].scatter(position[:index1,0],position[:index1,1], s = 15, c = cval1, cmap = 'plasma')
b = axs[0,1].scatter(position[N*100:index2,0],position[N*100:index2,1], s = 15, c = cval2, cmap = 'plasma')
c = axs[1,0].scatter(position[N*200:index3,0],position[N*200:index3,1], s = 15, c = cval3, cmap = 'plasma')
d = axs[1,1].scatter(position[N*350:index4,0],position[N*350:index4,1], s = 15, c = cval4, cmap = 'plasma')
```

### stellar merger simulation

```python
pos, vel, Ntot = collision(500, 400, 5, 0, 0, -.6, 0, 0)
```

```python
position, densities = main_SPH(pos, vel, Ntot, 2, .75, .1, 1, 6, 700, .04)
```
